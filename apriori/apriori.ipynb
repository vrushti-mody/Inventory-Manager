```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
#Gives you information about the dataset
dataset.info()
```

```python
#To display first 5 rows present in the dataset
dataset.head()
```

```python
#To check for duplicates
dataset.duplicated()
```

```python
#To delete duplicates
dataset.drop_duplicates(subset=None, inplace=True)
```

```python
transactions = []
for i in range(0, len(dataset)):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions, min_support = 0.00175, min_confidence = 0.5, min_lift = 2, min_length = 2)
```

```python
results = list(rules)
print(results)
```

```python
df = pd.DataFrame(results)
```

```python
df.to_csv(r'Output.csv', index = False)
```
